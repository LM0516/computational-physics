# Esercizio in Python

```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
def f(x):
    return np.exp(x)

def g(x, n):
    return sum((x**k)/math.factorial(k) for k in range(n))

def P(x, n):
    return x**(n + 1) / math.factorial(n + 1)
```

```python
x = np.linspace(0, 1)
n = 4
y_1 = f(x)
y_2 = g(x, n)
```

```python
delta = np.abs(y_1 - y_2)
poli = P(x, n)

plt.plot(x, delta)
plt.plot(x, poli, linestyle='dashed')
plt.xlabel('x')
plt.ylabel('|exp(x) - P_n(x)|')
```

```python
plt.plot(x, y_1)
plt.plot(x, y_2, linestyle='dashed')
plt.legend(['exp(x)', f'P_{n}(x)'])
```
